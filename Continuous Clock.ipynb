```python
#Continuous Clock
import turtle
h = turtle.Turtle()
h.pensize(4)
m = turtle.Turtle()
m.pensize(3)
s = turtle.Turtle()
h.hideturtle()
m.hideturtle()
s.hideturtle()
import time
class Clock:
    def circle(self):
        turtle.bgcolor('Sky blue')
        t = turtle.Turtle()
        t.hideturtle()
        t.speed(1000)
        t.penup()
        t.goto(0,-300)
        t.pendown()
        t.fillcolor('silver')
        t.begin_fill()
        t.circle(300)
        t.end_fill()
        t.lt(90)
        t.penup()
        t.fd(5)
        t.pendown()
        for i in range(6):
            t.pendown()
            t.fd(30)
            t.penup()
            t.fd(530)
            t.pendown()
            t.fd(30)
            t.lt(90)
            t.penup()
            for j in range(4):
                t.circle(290,6)
                t.lt(90)
                t.pendown()
                t.fd(10)
                t.penup()
                t.fd(570)
                t.pendown()
                t.fd(10)
                t.penup()
                t.lt(90)
            t.circle(290,6)
            t.lt(90)
    def hour(self):
        h.pencolor('silver')
        h.home()
        h.pencolor('black')
        h.lt(90)
        h.rt(int(self.h)*30)
        h.fd(100)
    def minute(self):
        m.pencolor('silver')
        m.home()
        m.pencolor('black')
        m.lt(90)
        m.rt(int(self.m)*6)
        m.fd(150)
    def second(self):
        s.pencolor('silver')
        s.home()
        s.pencolor('red')
        s.lt(90)
        s.rt(int(self.s)*6)
        s.fd(200)
class ContinuousClock(Clock):
    def gettime(self):
        super().circle()            
        self.h = time.strftime("%H")
        self.m = time.strftime("%M")
        self.s = time.strftime("%S")
        super().hour()
        super().minute()
        super().second()
        while True:
            if self.h != time.strftime("%H"):
                self.h = time.strftime("%H")
                super().hour()
            if self.m != time.strftime("%M"):
                self.m = time.strftime("%M")
                super().minute()
            if self.s != time.strftime("%S"):
                self.s = time.strftime("%S")
                super().second()
```

```python
import datetime
c = ContinuousClock()
c.gettime()
```
